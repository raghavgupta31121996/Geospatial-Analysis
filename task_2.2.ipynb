### Present a correlation plot between mean world population and mean per capita world urban population (from 1990 to 2010). Very briefly interpret the generated plot. In this question, you have to consider each year between 1990 and 2010 to find the mean population and mean per capita urban population.

```python
Total_P['Mean_Total'] = Total_P.iloc[:,34:55].mean(axis=1)
Total_P.head()
```

```python
Urban_P['correlation'] = Total_P['Mean_Total'].corr(Urban_P['MPC_WUP'])
Urban_P.head()
```

```python
result = pd.concat([Urban_P['MPC_WUP'], Total_P['Mean_Total']], axis=1)
result
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
ax = sns.heatmap(result.corr(),cmap='coolwarm',annot=True)
```

```python
plt.scatter(Urban_P['MPC_WUP'], Total_P['Mean_Total'])
plt.title('Correlation Plot from 1990 to 2010')
plt.xlabel('Mean per capita world urban population')
plt.ylabel('Mean population')
plt.show()
```

The above figures depicts the correlation heat map and a scatter plot between mean world population and mean per capita world urban 

population from 1990 to 2010. Correlation plot shows how a percent change in mean world population can affect the 

mean per capita world urban population. The plot exhibits that there is a negative correlation of -0.082 between 

the two. This means that on average as the total population increases the mean per capita world urban population 

decreases by 8.2%. It can be infered that between 1990 and 2010 there was a higher world rural population than 

world urban population on average which lead to increase in the total world population. As the mean world 

population rises, it becomes difficult to shift to urban areas and also for the country to provide facilities to everyone.
