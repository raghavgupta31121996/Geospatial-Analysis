#### Reading data files

Reading first World Bank data set, which includes total population. Since first 3 rows of the data set were of no use, therefore I skipped them.

```python
df1 = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_1637443.csv", skiprows=3)
df1.head()
```

Reading second World Bank data set, which includes urban population. Since first 3 rows of the data set were of no use, therefore I skipped them.

```python
df2 = pd.read_csv("API_SP.URB.TOTL_DS2_en_csv_v2_1868968.csv", skiprows=3)
df2.head()
```

#### Data Cleaning

dropna function in pandas help to remove the missing values from the dataset.

Removing coloumns from first data set having all NULL values.

```python
df1 = df1.dropna(how='all', axis=1)
df1.head()
```

Removing rows from first data set having any NULL values.

```python
df1 = df1.dropna(how='any', axis=0)
df1.head()
```

```python
df1.shape
```

Removing coloumns from second data set having all NULL values.

```python
df2 = df2.dropna(how='all', axis=1)
df2.head()
```

Removing rows from second data set having any NULL values.

```python
df2 = df2.dropna(how='any', axis=0)
df2.head()
```

```python
df2.shape
```

#### Inserting Geometric Data

#### Reading Geopandas inbuilt data

Since both World Bank data sets does not include the geometry of the countries, therefore reading inbuilt geopandas world data set. This data set includes the geometry and country code(iso_a3) of almost all the countries.

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
world.head()
```

Renaming coloumn iso_a3 to Country Code, same as the World Bank data set for further merging process.

```python
world = world.rename(columns={'iso_a3': 'Country Code'})
world.head()
```

Selecting columns from geopandas inbuilt dataset to merge geometry in both the World Bank datasets.

```python
world_shapes = world[['Country Code', 'geometry']]
world_shapes.head()
```

#### Adding geometry to World Bank datasets

Merge function in pandas help to merge two pandas datasets into a single dataset. 

Merging first dataset with geometric data

```python
df1_merged = df1.merge(world_shapes, on='Country Code', how='left')
df1_merged.head()
```

Merging second dataset with geometric data

```python
df2_merged = df2.merge(world_shapes, on='Country Code', how='left')
df2_merged.head()
```

#### Removing rows with no geometric value

```python
df1_merged = df1_merged.dropna(how='any', axis=0)
df1_merged.head()
```

```python
df2_merged = df2_merged.dropna(how='any', axis=0)
df2_merged.head()
```

#### Checking type of the dataframes

```python
type(df1_merged)
```

```python
type(df2_merged)
```

Since both dataframes are pandas dataframes, it is necessary to convert them into geopandas dataframes for further plotting process. 

#### Converting pandas dataframes to geopandas dataframes

```python
df1_gdf = gpd.GeoDataFrame(df1_merged, geometry = df1_merged.geometry, crs={'init': 'epsg:4326'})
df1_gdf.head()
```

```python
df2_gdf = gpd.GeoDataFrame(df2_merged, geometry = df2_merged.geometry, crs={'init': 'epsg:4326'})
df2_gdf.head()
```

#### Checking the type of dataframes again to see the success of previous step.

```python
type(df1_gdf)
```

```python
type(df2_gdf)
```

#### Saving both geopandas dataframes to shape files for further plotting process. 

```python
df1_gdf.to_file("Data_Total_Population.shp")
```

```python
df2_gdf.to_file("Data_Urban_Population.shp")
```

#### Reading first shape file.

```python
Total_P = gpd.read_file('Data_Total_Population.shp')
Total_P.head()
```

#### Plotting simple world map to check the size and coordinate reference system(CRS).

```python
Total_P.plot()
```

```python
Total_P.crs
```

#### Changing CRS to epsg = 3395 from epsg = 4326

```python
Total_P = Total_P.to_crs(epsg=3395)
Total_P.plot()
```

#### Reading second shape file and repeating the same steps as above.

```python
Urban_P = gpd.read_file('Data_Urban_Population.shp')
Urban_P.head()
```

```python
Urban_P.plot()
```

```python
Urban_P.crs
```

```python
Urban_P = Urban_P.to_crs(epsg=3395)
Urban_P.plot()
```

### Calculating world urban population per capita for years 1990, 2000 and 2010.

```python
Urban_P ['UBPC_1990'] = Urban_P['1990'] / Total_P['1990']
Urban_P ['UBPC_2000'] = Urban_P['2000'] / Total_P['2000']
Urban_P ['UBPC_2010'] = Urban_P['2010'] / Total_P['2010']
Urban_P.head()
```

```python
Urban_P.to_file("Data_Urban_Population.shp")
```

```python
Total_P ['UBPC_1990'] = Urban_P['1990'] / Total_P['1990']
Total_P ['UBPC_2000'] = Urban_P['2000'] / Total_P['2000']
Total_P ['UBPC_2010'] = Urban_P['2010'] / Total_P['2010']
Total_P.head()
```

```python
Total_P.to_file("Data_Total_Population.shp")
```

## Plotting choropleth map of World Urban Population Per Capita for year 1990.

matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)
divider = make_axes_locatable(ax)

cax = divider.append_axes("right", size="5%", pad=0.1)

Urban_P.plot(column='UBPC_1990', ax=ax, legend=True, cax=cax,legend_kwds={'label': "World Urban Population Per Capita Year - 1990"})
```
